```python
import pandas as pd
import matplotlib.pyplot as plt

the_office_df = pd.read_csv('datasets/office_episodes.csv')
the_office_df.head()
```

```python
colors = []

# color scheme reflecting the scaled ratings

for lab, row in the_office_df.iterrows():
    if row['scaled_ratings'] < 0.25:
        colors.append('red')
    elif 0.25 <= row['scaled_ratings'] < 0.50:
        colors.append('orange')
    elif 0.50 <= row['scaled_ratings'] < 0.75:
        colors.append('lightgreen')
    else:
        colors.append('darkgreen')
        
colors[:5]
```

```python
sizes = []

# sizing system for guests

for index, row in the_office_df.iterrows():
    if row['has_guests'] == False:
        sizes.append(25)
    else:
        sizes.append(250)

sizes[:15]
```

```python
fig = plt.rcParams['figure.figsize'] = [11, 7]
plt.style.use('fivethirtyeight')
fig = plt.figure()

plt.scatter(the_office_df['episode_number'], the_office_df['viewership_mil'], c=colors, s=sizes)

plt.title('Popularity, Quality, and Guest Appearances on the Office', size = 16)

plt.xlabel('Episode Number', size = 12)
plt.ylabel('Viewership (Millions)', size = 12)

# Show the plot
plt.show()
```

```python
highest_view = max(the_office_df['viewership_mil'])
most_watched = the_office_df.loc[the_office_df['viewership_mil'] == highest_view]

# who was in the most watched The Office episode
top_star = most_watched[['guest_stars']]
top_star
```
